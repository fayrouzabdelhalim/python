### **2.Find the 5 states with the lowest and highest `crime rates`**

```python
# find the 5 states with the lowest crime rates
lowest_crime_states = df[['State', 'Crime Rate']].sort_values(by='Crime Rate', ascending=True).head(5)

print(lowest_crime_states)
```

```python
# Create a bar plot to visualize the 5 states with the lowest crime rate

plt.figure(figsize=(10, 5))
sns.barplot(x='Crime Rate', y='State', data=lowest_crime_states, palette='viridis')
# Set title and labels for the plot
plt.title('Top 5 States with the Lowest Crime Rate', fontsize=14)
plt.xlabel('Crime Rate', fontsize=12)
plt.ylabel('State', fontsize=12)

# Display the plot
plt.show()

# Print your conclusions
print("Conclusion: The above plot shows the 5 states with the lowest crime rates. These states have significantly lower crime rates compared to others, indicating relatively safer regions.")
```

```python
# get the 5 states with the highest crime rates.
highest_crime_states = df[['State', 'Crime Rate']].sort_values(by='Crime Rate', ascending=False).head(5)
print(highest_crime_states)
```

```python
# Create a bar plot to visualize the 5 states with the Highest crime rate
plt.figure(figsize=(10, 5))
sns.barplot(x='Crime Rate', y='State', data=lowest_crime_states, palette="Reds")

# Set title and labels for the plot
plt.title('Top 5 States with the highestt Crime Rate', fontsize=14)
plt.xlabel('Crime Rate', fontsize=12)
plt.ylabel('State', fontsize=12)

# Display the plot
plt.show()

# Print your conclusions
print("Conclusion: The above plot highlights the 5 states with the highest crime rates using a red color scheme. These states have significantly higher crime rates, making them more dangerous compared to others.")
```

### **3.Find the sum of `Rape Cases` for each year**

```python
# Group by 'Year' and sum 'Rape Cases'
yearly_rape_cases = df.groupby('Year')['Rape Cases'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Rape Cases', data=yearly_rape_cases, marker='o', color='red', linewidth=2)

# Title and labels
plt.title('Trend of Total Rape Cases per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Rape Cases', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()

# Print conclusions
print("Conclusions: The line chart shows the trend of total rape cases over the years. An increasing trend may indicate a rise in crime, better reporting mechanisms, or both. A decreasing trend may reflect improved safety measures or underreporting. Further analysis is needed to interpret the exact reasons behind these trends.")
```

### **4.Find the correlation between `total crimes` and `population`**

```python
# Create a scatter plot with Population on the x-axis and Total Crimes on the y-axis
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Population', y='Total Crimes', data=df, color='green', alpha=0.6, edgecolor='black')
 #Log scale for better distribution (optional)
plt.xscale('log')  # Uncomment if population values are too large

# Set title and labels
plt.title('Relationship Between Population and Total Crimes', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Total Crimes', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()

# Print conclusions
#print("Conclusions: The scatter plot helps visualize the correlation between population size and total crimes. If there is an upward trend, it indicates that states with larger populations tend to have more total crimes. However, the density of points and any outliers should also be analyzed for better insights.")
```

### **5.Find out how the average of `crime clearance` rate vary across different `states`**

```python
# Group by 'State' and calculate the mean 'Crime Clearance Rate'
state_clearance_rates = df.groupby('State')['Crime Clearance Rate'].mean().reset_index()

# Sort in descending order
state_clearance_rates = state_clearance_rates.sort_values(by='Crime Clearance Rate', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Crime Clearance Rate', data=state_clearance_rates, palette='coolwarm')

# Set title and labels
plt.title('Average Crime Clearance Rate by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Crime Clearance Rate (%)', fontsize=12)

# Rotate x-axis labels to avoid overlap
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()

# Print conclusions
print("Conclusions: This bar plot shows the average crime clearance rate for each state. States with a higher clearance rate indicate effective law enforcement and case resolution, whereas lower rates may suggest inefficiencies or backlog in investigations.")
```

### **6.Find the Outliers of `abduction cases` in `western states`**

Note: wester states are 'Goa', 'Gujarat', 'Maharashtra', 'Rajasthan'

```python
# Define Western states
western_states = ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan']

# Filter dataset for Western states
western_df = df[df['State'].isin(western_states)]

# Plot the boxplot for 'Abduction Cases' in Western states
plt.figure(figsize=(8, 5))
sns.boxplot(x='State', y='Abduction Cases', data=western_df, palette='Reds')

# Set title and labels
plt.title('Outliers in Abduction Cases for Western States', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Abduction Cases', fontsize=12)

# Show the plot
plt.show()

# Print conclusions
print("Conclusions: The box plot identifies potential outliers in abduction cases for the Western states. Any points outside the whiskers of the boxplot are considered outliers, indicating unusually high or low abduction cases in that state.")
```

### **7.Find the trend in the number of `murder cases` over the `years`**

```python
 #Group by 'Year' and sum the 'Murder Cases'
murder_cases_trend = df.groupby('Year')['Murder Cases'].sum().reset_index()

# Create a line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Murder Cases', data=murder_cases_trend, marker='o', color='brown', linewidth=2)

# Set title and labels
plt.title('Trend of Murder Cases Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Murder Cases', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()

# Print conclusions
print("Conclusion: The trend shows how murder cases have changed over the years. If the line goes up, murder cases are increasing; if it goes down, they are decreasing.")
```

### **8.Find out which `state` has highest number of `arrest`**

```python

# Group by 'State' and sum the 'Arrests'
state_arrests = df.groupby('State')['Arrests'].sum().reset_index()

# Sort in descending order and pick the top 10
top_10_states = state_arrests.sort_values(by='Arrests', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(10, 5))
sns.barplot(x='State', y='Arrests', data=top_10_states, palette='Blues_r')

# Set title and labels
plt.title('Top 10 States with Highest Number of Arrests', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Arrests', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()

# Print conclusions
print("Conclusions: The bar plot shows the states with the highest number of arrests. The top statehas the most arrests, indicating higher crime rates or better law enforcement efficiency.")
```

### **9.Find the average `crime rate` across southern states over the `years`**

Note: southern states are 'Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry'

```python
# Define Southern states
southern_states = ['Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry']

# Filter dataset for Southern states
south_df = df[df['State'].isin(southern_states)]

# Group by 'Year' and calculate the average crime rate
avg_crime_rate = south_df.groupby('Year')['Crime Rate'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Crime Rate', data=avg_crime_rate, marker='o', color='black', linewidth=2)

# Set title and labels
plt.title('Average Crime Rate in Southern States Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Crime Rate', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.show()

# Print conclusions
print("Conclusions: The trend shows whether crime rates in Southern states are increasing or decreasing over time. If the line goes up, crime rates are rising; if it goes down, crime rates are improving.")
```

### **10.Find the Total number of `cases` for each crime type**

Note: cases are Rape cases, Abduction cases, Murder cases

```python
# Sum the cases for each crime type
crime_type_counts = df[['Rape Cases', 'Abduction Cases', 'Murder Cases']].sum()

# Plot the donut chart
plt.figure(figsize=(7, 7))
colors = ['#FF9999', '#66B3FF', '#99FF99']  # Custom colors for better visualization
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.gca().add_artist(plt.Circle((0, 0), 0.5, color='white'))  # Add white circle in the center to make it a donut


# Set the title

plt.title('Total Number of Cases for Each Crime Type', fontsize=14)

# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: largest segment represent most done crimes, to identify that Abduction cases are done the most , and there is a near percent between it and Rape cases")
```

### **12.Find the Correlation Matrix of Crime Data**

```python
# Filter data to keep only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the visual style
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Set the title
plt.title('Correlation Matrix of Crime Data', fontsize=14)

# Display the plot
plt.show()

# Print conclusions
print("Conclusions: The heatmap shows how different crime variables are related. A high positive correlation means crimes tend to occur together, while a negative correlation suggests an inverse relationship.")
```
